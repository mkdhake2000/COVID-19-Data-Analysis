Importing cowin_vaccine_data_districtwise.csv file

```python
import pandas as pd
dist_wise = pd.read_csv('cowin_vaccine_data_districtwise.csv', low_memory=False)
```

Import neighbor-districts-modified.json for getting the list of District Keys

```python
import json
with open('neighbor-districts-modified.json') as json_file:
    data = json.load(json_file)
dist_keys = list(data.keys())
```

Getting List of all Column names of the cowin_vaccine_data_districtwise.csv and getting only the dates from them

```python
lst = list(dist_wise)
lst
list1=lst[6:]
date_list = []
for i in range(len(list1)):
    if i%10 == 0:
        date_list.append(list1[i])
```

We only have to take data upto 14/08/2021. So, removing the later dates from the date list

```python
date_list=date_list[:date_list.index('14/08/2021')+1]
date_list
```

Replacing the NaN values with 0

```python
for j in date_list:
    dist_wise[j+'.5']=dist_wise[j+'.5'].fillna(0)
    dist_wise[j+'.6']=dist_wise[j+'.6'].fillna(0)
```

Importing data from the census file DDW_PCA0000_2011_Indiastatedist.xlsx

```python
census_data = pd.read_excel('DDW_PCA0000_2011_Indiastatedist.xlsx')
census_data
```

Getting the state code list for all states from district keys and taking the unique codes from them and sorting them

```python
states = [i.split('_')[0] for i in dist_keys]
states = set(states)
states = list(states)
states.sort()
states
```

Getting the list of state names from the state codes in the previous list using the state codes and state names from cowin_vaccine_data_districtwise.csv file

```python
state_keys = [0]*len(states)
for i in range(len(states)):
    k=list(dist_wise['State_Code']).index(states[i])
    state_keys[i]=dist_wise['State'][k]
state_keys
```

Changing Delhi to NCT OF DELHI, Spliting Dadra and Nagar Haveli and Daman and Diu as in census data and removing Ladakh and Telangana from state list since they are not there in the census data

```python
state_keys_census= state_keys[:]
state_keys_census[6]='NCT OF DELHI'
state_keys_census[7]='Dadra and Nagar Haveli'
state_keys_census.insert(8,'Daman and Diu')
state_keys_census.remove('Ladakh')
state_keys_census.remove('Telangana')
```

Replacing 'and' with '&' in states list

```python
for i in range(len(state_keys_census)):
    try:
        state_keys_census[i]=state_keys_census[i].replace(" and "," & ")
    except:
        print("No and")
```

```python
state_keys_census
```

Getting list of indices of states from census data

```python
state_in=[0]*len(state_keys_census)
for i in range(len(state_keys_census)):
    k=list(census_data['Name']).index(state_keys_census[i].upper())
    state_in[i]=census_data['State'][k]
state_in
```

Copying the name list of districts and states as it is from the census data

```python
dist_keys_census=census_data['Name'].copy()
```

Replacing double spaces with single spaces and stripping of front and end spaces of district keys from the census data

```python
for i in range(len(dist_keys_census)):
    if census_data['Level'][i]=='DISTRICT':
        dist_keys_census[i]=dist_keys_census[i].replace("  "," ")
        dist_keys_census[i]=dist_keys_census[i].strip()

dist_keys_census
```

Changing districts from the census data according to the cowin data

```python
for i in range(len(dist_keys_census)):
    if census_data['Level'][i]=='DISTRICT':
        if dist_keys_census[i] in list(change_dist.keys()):
            dist_keys_census[i]=change_dist[dist_keys_census[i]]
dist_keys_census
```

Getting indices of all districts from the cowin data in the census data

```python
dist_keys_index=[0]*len(dist_keys)
for i in range(len(dist_keys)):
    l=[index for index,el in enumerate(dist_keys_census) if el == dist_keys[i].split('_')[1]]
    l=[el for t,el in enumerate(l) if t%3==0]
    for ind in l:
        if dist_keys[i] not in dupl_lst:
            dist_keys_index[i]=census_data['District'][l[0]]
        else:
            j=census_data['State'][ind]
            k=state_keys_census.index(dat[dist_keys[i].split('_')[0]])
            if j==state_in[k]:
                dist_keys_index[i]=census_data['District'][ind]
dist_keys_index
```

Removing districts that are not common in census and cowin data and creating dictionary with keys as names of districts and values as indices of districts in the census data

```python
dist_keys_mod_dict={}
for i in range(len(dist_keys_index)):
    if dist_keys_index[i]!=0:
        dist_keys_mod_dict[dist_keys[i]]=dist_keys_index[i]
dist_keys_mod_dict
```

Getting list of keys of above dictionary i.e. the list of district keys common in both census and cowin data

```python
dist_keys_mod=dist_keys_mod_dict.keys()
dist_keys_mod=list(dist_keys_mod)
dist_keys_mod
```

Calculating vaccinated ratio of female to male for districts
Since the data is cumulative we only take data according to last date

```python
vccn_dist_ratio = [0]*len(dist_keys_mod)
for i in range(len(dist_keys_mod)):
    k=list(dist_wise['District_Key']).index(dist_keys_mod[i])
    vccn_dist_ratio[i]=int(dist_wise[date_list[-1]+'.6'][k])/int(dist_wise[date_list[-1]+'.5'][k])

vccn_dist_ratio
```

Getting the female and male population for each district from the census data

```python
dist_keys_mod_fpop = [0]*len(dist_keys_mod)
dist_keys_mod_mpop = [0]*len(dist_keys_mod)
for i in range(len(dist_keys_mod)):
    dist_keys_mod_fpop[i]=census_data['TOT_F'][list(census_data['District']).index(dist_keys_mod_dict[dist_keys_mod[i]])]
    dist_keys_mod_mpop[i]=census_data['TOT_M'][list(census_data['District']).index(dist_keys_mod_dict[dist_keys_mod[i]])]
```

Getting the population ratio of female population to male population for each district

```python
pop_dist_ratio=[0]*len(dist_keys_mod)
for i in range(len(dist_keys_mod)):
    pop_dist_ratio[i]=int(dist_keys_mod_fpop[i])/int(dist_keys_mod_mpop[i])
pop_dist_ratio
```

Getting the Vaccination-Population ratio for each district

```python
vccn_pop_dist_ratio=[0]*len(dist_keys_mod)
for i in range(len(vccn_pop_dist_ratio)):
    vccn_pop_dist_ratio[i]=vccn_dist_ratio[i]/pop_dist_ratio[i]
vccn_pop_dist_ratio
```

Combining Columns into a dataframe

```python
data_final = {'districtid':dist_keys_mod, 'vaccinationratio':vccn_dist_ratio,'populationratio':pop_dist_ratio,'ratioofratios':vccn_pop_dist_ratio}
data_dist = pd.DataFrame(data_final)
data_dist
```

Sorting the dataframe based on the final ratio

```python
data_dist=data_dist.sort_values('ratioofratios')
```

Getting list of state codes from the modified district keys list i.e. after taking the common from both census and cowin data

```python
states_mod = [i.split('_')[0] for i in dist_keys_mod]
states_mod = set(states)
states_mod = list(states)
states_mod.sort()
states_mod
```

Getting the names of states from cowin data usig the state codes calculated from previous cell

```python
state_keys_mod = [0]*len(states_mod)
for i in range(len(states_mod)):
    k=list(dist_wise['State_Code']).index(states_mod[i])
    state_keys_mod[i]=dist_wise['State'][k]
state_keys_mod
```

Getting the female and male population for each state. We took the sum of the respective populations of each district which is a part of that state

```python
states_mod_fpop=[0]*len(states_mod)
states_mod_mpop=[0]*len(states_mod)
for i in range(len(states_mod)):
    tmpf=0
    tmpm=0
    l=[ind for ind,el in enumerate(dist_keys_mod) if el.split('_')[0] == states_mod[i]]
    for j in l:
        tmpf=tmpf+dist_keys_mod_fpop[j]
        tmpm=tmpm+dist_keys_mod_mpop[j]
    states_mod_fpop[i]=tmpf
    states_mod_mpop[i]=tmpm
```

Getting the female and male vaccinations for each state. We took the sum of the respective vaccinations of each district which is a part of that state

```python
states_mod_fvccn=[0]*len(states_mod)
states_mod_mvccn=[0]*len(states_mod)
for i in range(len(states_mod)):
    tmpf=0
    tmpm=0
    l=[ind for ind,el in enumerate(dist_keys_mod) if el.split('_')[0] == states_mod[i]]
    for j in l:
        tmpf=tmpf+int(dist_wise[date_list[-1]+'.6'][list(dist_wise['District_Key']).index(dist_keys_mod[j])])
        tmpm=tmpm+int(dist_wise[date_list[-1]+'.5'][list(dist_wise['District_Key']).index(dist_keys_mod[j])])
    states_mod_fvccn[i]=tmpf
    states_mod_mvccn[i]=tmpm
```

Getting the female to male vaccination ratio for each state

```python
vccn_state_ratio=[0]*len(states_mod)
for i in range(len(states_mod)):
    vccn_state_ratio[i]=int(states_mod_fvccn[i])/int(states_mod_mvccn[i])

vccn_state_ratio
```

Getting the female to male population ratio for each state

```python
pop_state_ratio=[0]*len(states_mod)
for i in range(len(states_mod)):
    pop_state_ratio[i]=int(states_mod_fpop[i])/int(states_mod_mpop[i])
pop_state_ratio
```

Getting the Vaccination-Population ratio for each state

```python
vccn_pop_state_ratio=[0]*len(states_mod)
for i in range(len(states_mod)):
    vccn_pop_state_ratio[i]=vccn_state_ratio[i]/pop_state_ratio[i]
vccn_pop_state_ratio
```

Combining the columns to dataframe

```python
data_final = {'stateid':state_keys_mod, 'vaccinationratio':vccn_state_ratio,'populationratio':pop_state_ratio,'ratioofratios':vccn_pop_state_ratio}
data_state = pd.DataFrame(data_final)
data_state
```

Sorting the dataframe based on the final ratio

```python
data_state=data_state.sort_values('ratioofratios')
```

Exporting dataframe to state-vaccination-population-ratio.csv

```python
data_state.to_csv('state-vaccination-population-ratio.csv',index=False)
```
