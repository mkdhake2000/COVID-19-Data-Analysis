Importing cowin_vaccine_data_districtwise.csv file

```python
import pandas as pd
dist_wise = pd.read_csv('cowin_vaccine_data_districtwise.csv', low_memory=False)
```

Importing math and datetime libraries

```python
census_data = pd.read_excel('DDW_PCA0000_2011_Indiastatedist.xlsx')
census_data
```

Import neighbor-districts-modified.json for getting the list of District Keys

```python
import json
with open('neighbor-districts-modified.json') as json_file:
    data = json.load(json_file)
dist_keys = list(data.keys())
```

Getting List of all Column names of the cowin_vaccine_data_districtwise.csv and getting only the dates from them

```python
lst = list(dist_wise)
lst
list1=lst[6:]
date_list = []
for i in range(len(list1)):
    if i%10 == 0:
        date_list.append(list1[i])
```

We only have to take data upto 14/08/2021. So, removing the later dates from the date list

```python
date_list=date_list[:date_list.index('14/08/2021')+1]
date_list
```

Getting the state code list for all states from district keys and taking the unique codes from them and sorting them

```python
states = [i.split('_')[0] for i in dist_keys]
states = set(states)
states = list(states)
states.sort()
states
```

Getting the list of state names from the state codes in the previous list using the state codes and state names from cowin_vaccine_data_districtwise.csv file

```python
state_keys = [0]*len(states)
for i in range(len(states)):
    k=list(dist_wise['State_Code']).index(states[i])
    state_keys[i]=dist_wise['State'][k]
state_keys
```

Changing Delhi to NCT OF DELHI, Spliting Dadra and Nagar Haveli and Daman and Diu as in census data and removing Ladakh and Telangana from state list since they are not there in the census data

```python
state_keys_census= state_keys[:]
state_keys_census[6]='NCT OF DELHI'
state_keys_census[7]='Dadra and Nagar Haveli'
state_keys_census.insert(8,'Daman and Diu')
state_keys_census.remove('Ladakh')
state_keys_census.remove('Telangana')
```

Replacing and with & in states list

```python
for i in range(len(state_keys_census)):
    try:
        state_keys_census[i]=state_keys_census[i].replace(" and "," & ")
    except:
        print("No and")
```

Changing districts from the census data according to the cowin data

```python
for i in range(len(dist_keys_census)):
    if census_data['Level'][i]=='DISTRICT':
        if dist_keys_census[i] in list(change_dist.keys()):
            dist_keys_census[i]=change_dist[dist_keys_census[i]]
dist_keys_census
```

Getting indices of all districts from the cowin data in the census data

```python
dist_keys_index=[0]*len(dist_keys)
for i in range(len(dist_keys)):
    l=[index for index,el in enumerate(dist_keys_census) if el == dist_keys[i].split('_')[1]]
    l=[el for t,el in enumerate(l) if t%3==0]
    for ind in l:
        if dist_keys[i] not in dupl_lst:
            dist_keys_index[i]=census_data['District'][l[0]]
        else:
            j=census_data['State'][ind]
            k=state_keys_census.index(dat[dist_keys[i].split('_')[0]])
            if j==state_in[k]:
                dist_keys_index[i]=census_data['District'][ind]
dist_keys_index
```

Removing districts that are not common in census and cowin data and creating dictionary with keys as names of districts and values as indices of districts in the census data

```python
dist_keys_mod_dict={}
for i in range(len(dist_keys_index)):
    if dist_keys_index[i]!=0:
        dist_keys_mod_dict[dist_keys[i]]=dist_keys_index[i]
dist_keys_mod_dict
```

Getting list of keys of above dictionary i.e. the list of district keys common in both census and cowin data

```python
dist_keys_mod=dist_keys_mod_dict.keys()
dist_keys_mod=list(dist_keys_mod)
dist_keys_mod
```

Getting the total population for each district from the census data

```python
dist_keys_mod_pop = [0]*len(dist_keys_mod)
for i in range(len(dist_keys_mod)):
    dist_keys_mod_pop[i]=census_data['TOT_P'][list(census_data['District']).index(dist_keys_mod_dict[dist_keys_mod[i]])]
```

Getting the number of dose 1 vaccinated people from cowin data that is yet to be vaccinated with even dose 1
We have to consider the rate only for the last week with that week ending on 14/08/2021.
So, we have to take date for 14/08/2021 and substract the data for 07/08/2021, since the data is cumulative

```python
dist_dose1=[0]*len(dist_keys_mod)
for i in range(len(dist_keys_mod)):
    k=list(dist_wise['District_Key']).index(dist_keys_mod[i])
    dist_dose1[i]=int(dist_wise[date_list[-1]+'.3'][k])-int(dist_wise[date_list[-8]+'.3'][k])
```

Getting list of state codes from the modified district keys list i.e. after taking the common from both census and cowin data

```python
states_mod = [i.split('_')[0] for i in dist_keys_mod]
states_mod = set(states)
states_mod = list(states)
states_mod.sort()
states_mod
```

Getting the names of states from cowin data usig the state codes calculated from previous cell

```python
state_keys_mod = [0]*len(states_mod)
for i in range(len(states_mod)):
    k=list(dist_wise['State_Code']).index(states_mod[i])
    state_keys_mod[i]=dist_wise['State'][k]
state_keys_mod
```

Getting the total population, dose 1 vacccinated number of people and remaining population yet to take dose 1 vaccine for each state. We took the sum of the respective populations of each district which is a part of that state

```python
states_mod_tpop=[0]*len(states_mod)
states_mod_dose1=[0]*len(states_mod)
states_mod_rem=[0]*len(states_mod)
for i in range(len(states_mod)):
    tmpt=0
    tmpd=0
    l=[ind for ind,el in enumerate(dist_keys_mod) if el.split('_')[0] == states_mod[i]]
    for j in l:
        tmpt=tmpt+dist_keys_mod_pop[j]
        tmpd=tmpd+dist_dose1[j]
    states_mod_tpop[i]=tmpt
    states_mod_dose1[i]=tmpd/7
    states_mod_rem[i]=tmpt-tmpd
```

Getting the dates till when the whole population will get vaccinated at least by dose 1 for all states

```python
states_date_end = [0]*len(states_mod)
curr_date=date_list[-1]
curr_date_tmp=datetime.datetime.strptime(curr_date, "%d/%m/%Y")
for i in range(len(states_mod)):
    days_req=math.ceil(states_mod_rem[i]/states_mod_dose1[i])
    states_date_end[i]=curr_date_tmp+datetime.timedelta(days=days_req)
    states_date_end[i]=states_date_end[i].strftime("%d/%m/%Y")
```

Combining columns to dataframe

```python
data_final = {'stateid':state_keys_mod, 'populationleft':states_mod_rem,'rateofvaccination':states_mod_dose1,'date':states_date_end}
data_state = pd.DataFrame(data_final)
data_state
```

Exporting dataframe to complete-vaccination.csv

```python
data_state.to_csv('complete-vaccination.csv',index=False)
```
