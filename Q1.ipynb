```python
for i in range(len(dist_keys)):
    dist_keys[i] = dist_keys[i].split("/")[0]
```

```python
for i in data.keys():
    for j in range(len(data[i])):
        data[i][j] = data[i][j].split("_district")[0]
        data[i][j] = data[i][j].replace("_"," ")
        for k in range(len(dupl_dist)):
            if data[i][j] == dupl_dist[k]:
                data[i][j] = dupl_dist_repl[k]
dist_values = list(data.values())
dist_values
```

```python
lst = []
for i in range(len(dist_keys)):
        
    if '_' not in dist_keys[i]:
        lst.append(dist_keys[i])
print(lst)
```

Removing districts from the above found list from the district keys list

```python
for i in dist_keys:
    for j in lst:
            if i is j:
                data.pop(j)
                
dist_keys = list(data.keys())
dist_keys
```

Importing collections.OrderedDict to sort the dictionary with respect to it's keys

```python
from collections import OrderedDict
```

Removing districts from the above found list from the district values list

And Merging district keys and district values into a dictionary

```python
for i in dist_values:
    for k in list(i):
        if k in lst:
            ind=dist_values.index(i)
            dist_values[ind].remove(k)
data = dict(zip(dist_keys, dist_values))
data1=OrderedDict(sorted(data.items()))
data1
```
