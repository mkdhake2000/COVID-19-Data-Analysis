Importing cowin_vaccine_data_districtwise.csv

```python
import pandas as pd
dist_wise = pd.read_csv('cowin_vaccine_data_districtwise.csv', low_memory=False)
```

Import neighbor-districts-modified.json for getting the list of District Keys

```python
import json
with open('neighbor-districts-modified.json') as json_file:
    data = json.load(json_file)
dist_keys = list(data.keys())
```

Getting List of all Column names of the cowin_vaccine_data_districtwise.csv and getting only the dates from them

```python
lst = list(dist_wise)
lst
list1=lst[6:]
date_list = []
for i in range(len(list1)):
    if i%10 == 0:
        date_list.append(list1[i])
```

We only have to take data upto 14/08/2021. So, removing the later dates from the date list

```python
date_list=date_list[:date_list.index('14/08/2021')+1]
date_list
```

Replacing the NaN values with 0

```python
for j in date_list:
    dist_wise[j+'.8']=dist_wise[j+'.8'].fillna(0)
    dist_wise[j+'.9']=dist_wise[j+'.9'].fillna(0)
```

Getting the number of covaxin vaccinated people and covishield vaccinated people.
Since, the data is cumulative we only consider the data from the last date .i.e. here 14/08/2021

```python
cova_dist=[0]*len(dist_keys)
covi_dist=[0]*len(dist_keys)
for i in range(len(dist_keys)):
    k=list(dist_wise['District_Key']).index(dist_keys[i])
    cova_dist[i]=dist_wise[date_list[-1]+'.8'][k]
    covi_dist[i]=dist_wise[date_list[-1]+'.9'][k]
```

Taking the ratio of covishield vaccinated people to covaxin vaccinated people for all districts.
Where, Covaxin vaccinated people count is zero, we get dividingbyzero exception. So, implementing a try except block to catch dividingbyzero exception and setting the value that is greatest of all values of the ratio i.e. 9999999999 so that it becomes easy to sort the values according to the ratio

```python
covi_cova_dist=[0]*len(dist_keys)
for i in range(len(dist_keys)):
    try:
        covi_cova_dist[i]=int(covi_dist[i])/int(cova_dist[i])
    except:
        covi_cova_dist[i]=9999999999
```

```python
covi_cova_dist
```

Combining columns to a dataframe

```python
data_final = {'districtid':dist_keys, 'vaccineratio':covi_cova_dist}
data_dist = pd.DataFrame(data_final)
data_dist
```

Sorting the dataframe with respect to the vaccineratio column

```python
data_dist=data_dist.sort_values('vaccineratio')
```

Replacing 9999999999 with 'N/A'

```python
for i in range(len(data_dist)):
    if data_dist['vaccineratio'][i]==9999999999:
        data_dist['vaccineratio'][i]= 'N/A'
```

```python
data_dist
```

Exporting the modified dataframe to district-vaccine-type-ratio.csv

```python
data_dist.to_csv('district-vaccine-type-ratio.csv',index=False)
```

Getting the state code list for all states from district keys and taking the unique codes from them and sorting them

```python
states = [i.split('_')[0] for i in dist_keys]
states = set(states)
states = list(states)
states.sort()
states
```

Getting the list of state names from the state codes in the previous list using the state codes and state names from cowin_vaccine_data_districtwise.csv file

```python
state_keys = [0]*len(states)
for i in range(len(states)):
    k=list(dist_wise['State_Code']).index(states[i])
    state_keys[i]=dist_wise['State'][k]
state_keys
```

Getting the number of covaxin vaccinated people and covishield vaccinated people.
We find that data for each state by adding the data for all districts which are a part of that state

```python
cova_state=[0]*len(states)
covi_state=[0]*len(states)
for i in range(len(states)):
    tmpca=0
    tmpci=0
    l=[ind for ind,el in enumerate(dist_keys) if el.split('_')[0] == states[i]]
    for j in l:
        tmpca=tmpca+int(cova_dist[j])
        tmpci=tmpci+int(covi_dist[j])
    cova_state[i]=tmpca
    covi_state[i]=tmpci
```

Taking the ratio of covishield vaccinated people to covaxin vaccinated people for all states.
Where, Covaxin vaccinated people count is zero, we get dividingbyzero exception. So, implementing a try except block to catch dividingbyzero exception and setting the value that is greatest of all values of the ratio i.e. 9999999999 so that it becomes easy to sort the values according to the ratio

```python
covi_cova_state=[0]*len(states)
for i in range(len(states)):
    try:
        covi_cova_state[i]=int(covi_state[i])/int(cova_state[i])
    except:
        covi_cova_state[i]=9999999999
```

```python
covi_cova_state
```

Combining the columns into a dataframe

```python
data_final = {'stateid':state_keys, 'vaccineratio':covi_cova_state}
data_state = pd.DataFrame(data_final)
data_state
```

Sorting the dataframe with respect to the vaccineratio column

```python
data_state=data_state.sort_values('vaccineratio')
```

Replacing 9999999999 with 'N/A'

```python
for i in range(len(data_state)):
    if data_state['vaccineratio'][i]==9999999999:
        data_state['vaccineratio'][i]= 'N/A'
```

```python
data_state
```

Exporting the modified dataframe to state-vaccine-type-ratio.csv

```python
data_state.to_csv('state-vaccine-type-ratio.csv',index=False)
```

Getting the number of covaxin vaccinated people and the covishield vaccinated people overall. We find that by adding the respective data from all states

```python
cova_overall=0
covi_overall=0
for i in range(len(states)):
    covi_overall=covi_overall+covi_state[i]
    cova_overall=cova_overall+cova_state[i]
```

Taking the ratio of covishield vaccinated people to covaxin vaccinated people overall

```python
covi_cova_overall=int(covi_overall)/int(cova_overall)
```

```python
covi_cova_overall
```

Combining the columns into a dataframe

```python
data_final = {'Overall':['Overall'], 'vaccineratio':[covi_cova_overall]}
data_overall = pd.DataFrame(data_final)
data_overall
```

Exporting dataframe to overall-vaccine-type-ratio.csv

```python
data_overall.to_csv('overall-vaccine-type-ratio.csv',index=False)
```
